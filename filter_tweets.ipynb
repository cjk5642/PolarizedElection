```python
import pandas as pd
import numpy as np
import os
```

```python
folder = r".\Data\Tweets\biden_election"
files = [os.path.join(folder,i) for i in os.listdir(folder) if i.endswith('.csv')]

states = pd.read_csv(r".\Data\Misc\states.csv")
states = [states.iloc[i, :] for i in range(len(states))]
states = [j for i in states for j in i]

df = pd.DataFrame(columns = pd.read_csv(files[0]).columns)

for fd in files:
    frame = pd.read_csv(fd)
    frame = frame.dropna(axis = 0, subset = ['location'])
    df = df.append(frame, ignore_index = True)

rows = []
for row in range(len(df)):
    for state in states:
        if str(state) in str(df.iloc[row,2]):
            rows.append(row)
df = df.loc[rows,:].drop_duplicates().reset_index(drop = True)
```

```python
dfs = df

users = list(set(dfs.username))

data = pd.DataFrame(columns = ['username','location','all_tweet'])
for i,u in enumerate(users):
    d = dfs[dfs['username'] == str(u)]
    words = []
    for row in range(len(d)):
        word = str(d.iloc[row, -2]).strip().split(' ')
        words += word
    rows = [list(set(d.username))[0], list(set(d.location))[0], list(set(words))]
    data = data.append({'username':rows[0], 'location':rows[1], 'all_tweet':rows[2]}, ignore_index = True)

display(data)
```
